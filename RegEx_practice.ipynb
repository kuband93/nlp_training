```
#import library
import re
```

```
string = "some string"
regex = r"\w"

re.findall(regex, string)
```

```
regex = r"\w+"
re.findall(regex, string)
```

```
regex1 = r"\W"
regex2 = r"\W+"
print(re.findall(regex1, string), re.findall(regex1, string))
```

```
string = "That's some nice string!"
print(re.findall(regex1, string), re.findall(regex1, string))
```

```
#testing RegEx most common expressions

regex = r"\w\w\w"
string = "--->abc<----"
print(re.findall(regex, string))

regex = r"\W\Wabc"
string = "Password is: !#abc"
print(re.findall(regex, string))

regex = r"\d\d\d-\d\d\d-\d\d\d"
string = "My number is: 789-241-243"
print(re.findall(regex, string))

regex = r"\D\D\D"
string = "Hi!"
print(re.findall(regex, string))

regex = r"\w\w\s\d\d"
string = "ID 89"
print(re.findall(regex, string))

regex = r"\S\S\s\S"
string = "Hi !"
print(re.findall(regex, string))

regex = r"\w+@\S+"
string = "jakub@gmail.com"
print(re.findall(regex, string))

regex = r"[a-z]+"
string = "allgood"
print(re.findall(regex, string))

regex = r".*"
string = "997 trouble!"
print(re.findall(regex, string))
```

```
#Exersie 1 - find correct emails

string = "Hi my name is John Clyde. My business email is: john.clyde@tesla.com. My personal email is: john1993@gmail.com. Funny story last week I sign up your newsletter as I entered wrong email - john.clyde!@tesla.com"
regex = r"[a-z]\S+[a-z0-9]@\w+.com"

re.findall(regex,string)
```

```
#Exercise 2 - split the sentences on special signs at theri end

string = "How to split the senteces? Maybe it's not as difficult as it looks. Let's try!"
regex = r'[?.!]'

re.split(regex,string)
```

```
#Exerscie 3 - run the code and understand the difference between search and match methods

string = "abc - let's find it!"
regex = r"abc"
print("Match method: "+ str(re.match(regex,string)) + "\nSearch method: " + str(re.search(regex,string)))


string = "let's find abc!"
regex = r'abc'
print("\nMatch method: " + str(re.match(regex,string)) + "\nSearch method: " + str(re.search(regex,string)))
```

```
#Excersie 4 - substitute all . for !

string =  "Hi, my name is Jakub. Hope you are doing great."
regex = r'\.'
print(re.sub(regex, "!", string))
```
